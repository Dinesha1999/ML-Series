```python
#import the libraries
import numpy as np
import pandas as pd
```

```python
#open the dataset
data = pd.read_csv('kyphosis.csv')
```

```python
#if need to check some data.here the label is kyphosis,and others are features
data.head()
```

```python
#looking the shape(raws & columns) of dataset
data.shape
```

```python
#to check null/missing values
data.info()
```

```python
#now spit the data into pieces.Take the label value as Y value & taking other columns as like X(values wich is given to train)
#by removing kyphosis column/droping column can be define X
x = data.drop('Kyphosis',axis = 1) #column/axis
```

```python
x.head()
```

```python
#define the Y taking kyphosis column
y = data['Kyphosis']
y.head() #look forY values
```

```python
#devide the data set to train & test
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.3) # as giving x and y values as the input & take 30% data to test  
```

```python
x_train.shape
#train the dataset using 56 records
```

```python
x_test.shape
```

## Decisin Tree

```python
#Now train this model using DECISION TREE
from sklearn.tree import DecisionTreeClassifier
#create the object called model
model = DecisionTreeClassifier()
#calling fit method through object to train and give itto traing ndataset
model.fit(x_train,y_train)
```

```python
#now the model is trained/finished according above implement and model can predict new values
predict = model.predict(x_test)
predict
```

```python
#to get actual values according above prediction
y_test
```

```python
#checking performance/accuracy of the model with giving actual y values & predicted values by model
from sklearn.metrics import accuracy_score
accuracy_score(y_test,predict)
```

```python
#Confusion matrix/predicted values as wrong 1+5
from sklearn.metrics import confusion_matrix
confusion_matrix(y_test,predict)
```
